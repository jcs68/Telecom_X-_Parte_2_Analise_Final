```python
# Carregar os dados tratados
df = pd.read_csv("/content/dados_tratados.csv")

# 1ï¸âƒ£ Remover colunas com alta cardinalidade (ex: cÃ³digos ou identificadores)
card_limit = len(df) * 0.9
altamente_cardinais = [col for col in df.columns if df[col].nunique() > card_limit]
df.drop(columns=altamente_cardinais, inplace=True)

# 2ï¸âƒ£ Remover colunas com valor constante
constantes = [col for col in df.columns if df[col].nunique() == 1]
df.drop(columns=constantes, inplace=True)

# 3ï¸âƒ£ Remover colunas duplicadas (estrutura idÃªntica)
df = df.loc[:, ~df.T.duplicated()]

# 4ï¸âƒ£ Remover colunas com muitos nulos (ex: mais de 40%)
nulo_limite = 0.4
nulos = df.columns[df.isnull().mean() > nulo_limite]
df.drop(columns=nulos, inplace=True)

# 5ï¸âƒ£ Verificar correlaÃ§Ãµes fortes (redundÃ¢ncia numÃ©rica)
correlacoes = df.corr(numeric_only=True)
correlacoes_altas = correlacoes[(correlacoes > 0.9) & (correlacoes < 1.0)].stack()
print("âš ï¸ VariÃ¡veis com alta correlaÃ§Ã£o:\n", correlacoes_altas)

# Resultado final: visualizar colunas restantes
print("ðŸ§® Colunas apÃ³s limpeza:\n", df.columns.tolist())
```

### Encoding

```python
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import OneHotEncoder

# Identificar colunas categÃ³ricas (objetos/textos)
colunas_categoricas = df.select_dtypes(include="object").columns.tolist()

# Criar o transformador com OneHotEncoder
one_hot_enc = make_column_transformer(
    (OneHotEncoder(handle_unknown='ignore'), colunas_categoricas),
    remainder='passthrough'  # mantÃ©m colunas numÃ©ricas sem alteraÃ§Ãµes
)

# Aplicar transformaÃ§Ã£o aos dados
dados_transformados = one_hot_enc.fit_transform(df)

# Criar DataFrame com os novos nomes de colunas
df_encoded = pd.DataFrame(dados_transformados, columns=one_hot_enc.get_feature_names_out())

# Visualizar o resultado
df_encoded.head()
```

## Balanceamento de Classes (opcional )

Caso queira aprofundar a anÃ¡lise, aplique tÃ©cnicas de balanceamento como undersampling ou oversampling. Em situaÃ§Ãµes de forte desbalanceamento, ferramentas como o SMOTE podem ser Ãºteis para gerar exemplos sintÃ©ticos da classe minoritÃ¡ria.


```python
from imblearn.pipeline import Pipeline  # â¬…ï¸ pipeline da imblearn!
from imblearn.over_sampling import SMOTE
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score

# Pipeline corrigido
pipeline = Pipeline([
    ('smote', SMOTE(random_state=42)),
    ('model', RandomForestClassifier(random_state=42))
])

# Separar variÃ¡veis
X = df_encoded.drop(columns=['remainder__Cancelamento'])
y = df_encoded['remainder__Cancelamento']

# ValidaÃ§Ã£o cruzada
scores = cross_val_score(pipeline, X, y, cv=5, scoring='f1')
print("ðŸ“Š F1 MÃ©dio:", scores.mean())

```

```python
from imblearn.combine import SMOTETomek
from imblearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score

# Substituindo SMOTE por SMOTETomek no pipeline
pipeline = Pipeline([
    ('smotetomek', SMOTETomek(random_state=42)),
    ('model', RandomForestClassifier(random_state=42))
])

# Separando X e y
X = df_encoded.drop(columns=['remainder__Cancelamento'])
y = df_encoded['remainder__Cancelamento']

# Avaliando desempenho
scores = cross_val_score(pipeline, X, y, cv=5, scoring='f1')
print("ðŸ“Š F1 MÃ©dio com SMOTETomek:", scores.mean())
```

```python
from sklearn.model_selection import GridSearchCV


# 1. Pipeline com balanceamento + modelo
pipeline = Pipeline([
    ('smotetomek', SMOTETomek(random_state=42)),
    ('model', RandomForestClassifier(random_state=42))
])

# 2. Grade de hiperparÃ¢metros
param_grid = {
    'model__n_estimators': [100, 200, 300],
    'model__max_depth': [None, 5, 10, 15],
    'model__min_samples_split': [2, 5],
    'model__min_samples_leaf': [1, 2]
}

# 3. Rodando o GridSearch
grid_search = GridSearchCV(pipeline, param_grid, scoring='f1', cv=5, n_jobs=-1)
grid_search.fit(X, y)

# 4. Resultados
print("ðŸ† Melhor F1-score:", grid_search.best_score_)
print("ðŸ” Melhor combinaÃ§Ã£o de parÃ¢metros:")
print(grid_search.best_params_)
```

### AnÃ¡lise do resultado F1-score com GridSearchCV:

O modelo atingiu um F1-score mÃ©dio de 0.6167 com os melhores hiperparÃ¢metros encontrados â€” isso representa uma melhoria significativa em relaÃ§Ã£o aos testes anteriores com SMOTE (â‰ˆ0.538) e SMOTETomek simples (â‰ˆ0.550). O ganho acumulado mostra que o modelo agora captura melhor os clientes que cancelaram e estÃ¡ mais equilibrado entre precisÃ£o e recall. Vamos destrinchar:

### O que cada ajuste contribuiu:

*  **max_depth = 5:** reduz a complexidade da Ã¡rvore, evita divisÃµes excessivas.a

*   **min_samples_leaf = 2:** garante que folhas tenham ao menos 2 amostras, evitando ruÃ­do

*   **min_samples_split = 2:** permite dividir nÃ³s com pelo menos 2 amostras â€” Ãºtil para capturar padrÃµes sutis.

*   **n_estimators = 200:** mais Ã¡rvores aumentam estabilidade sem sobrecarregar.

### InterpretaÃ§Ã£o prÃ¡tica:

O modelo estÃ¡ mais equilibrado: detecta evasores sem sacrificar acertos entre clientes ativos.

Esse desempenho jÃ¡ Ã© considerado bom para problemas reais com dados desbalanceados (~26.6% de evasÃ£o).

## SeparaÃ§Ã£o de Dados

Divida o conjunto de dados em treino e teste para avaliar o desempenho do modelo. Uma divisÃ£o comum Ã© 70% para treino e 30% para teste, ou 80/20, dependendo do tamanho da base de dados.

```python
from sklearn.model_selection import train_test_split

# Separar features (X) e target (y)
X = df_encoded.drop(columns=['remainder__Cancelamento'])
y = df_encoded['remainder__Cancelamento']

# DivisÃ£o treino/teste â€” 70%/30%
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)
```

## CriaÃ§Ã£o de Modelos

Crie pelo menos dois modelos diferentes para prever a evasÃ£o de clientes.

Um modelo pode exigir normalizaÃ§Ã£o, como RegressÃ£o LogÃ­stica ou KNN.

O outro modelo pode nÃ£o exigir normalizaÃ§Ã£o, como Ãrvore de DecisÃ£o ou Random Forest.

ðŸ’¡ A escolha de aplicar ou nÃ£o a normalizaÃ§Ã£o depende dos modelos selecionados. Ambos os modelos podem ser criados sem normalizaÃ§Ã£o, mas a combinaÃ§Ã£o de modelos com e sem normalizaÃ§Ã£o tambÃ©m Ã© uma opÃ§Ã£o.

Justifique a escolha de cada modelo e, se optar por normalizar os dados, explique a necessidade dessa etapa.

### Modelo 1: RegressÃ£o LogÃ­stica (com normalizaÃ§Ã£o)

ðŸ” Justificativa:
A RegressÃ£o LogÃ­stica Ã© um modelo linear e interpretÃ¡vel, Ã³timo para identificar relaÃ§Ãµes entre variÃ¡veis e a probabilidade de cancelamento. Ela se destaca em cenÃ¡rios com dados limpos e balanceados, como os seus apÃ³s SMOTETomek.

### ðŸ§ª Por que exige normalizaÃ§Ã£o?

Modelos lineares sÃ£o sensÃ­veis Ã  escala das variÃ¡veis.

Features com magnitudes maiores podem dominar o processo de aprendizado.

A normalizaÃ§Ã£o (ou padronizaÃ§Ã£o) garante que cada feature contribua de forma justa para o modelo.

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression

# Modelo com normalizaÃ§Ã£o
pipeline_lr = Pipeline([
    ('scaler', StandardScaler()),
    ('model', LogisticRegression(max_iter=1000, random_state=42))
])

pipeline_lr.fit(X_train, y_train)
```

### **ðŸŒ³ Modelo 2: Random Forest (sem normalizaÃ§Ã£o)**

**ðŸ” Justificativa:**

Random Forest Ã© um modelo baseado em Ã¡rvores de decisÃ£o. Ideal para lidar com dados mistos (numÃ©ricos e categÃ³ricos), lidar com outliers e capturar interaÃ§Ãµes nÃ£o lineares entre variÃ¡veis.

**âœ… Vantagens:**

NÃ£o exige normalizaÃ§Ã£o porque usa medidas de impureza (ex.: Gini) e nÃ£o depende de distÃ¢ncias.

JÃ¡ demonstrou bom desempenho nos seus testes anteriores com F1-score superior a 0.61.

```python
from sklearn.ensemble import RandomForestClassifier

model_rf = RandomForestClassifier(n_estimators=200, max_depth=5, min_samples_split=2, min_samples_leaf=2, random_state=42)
model_rf.fit(X_train, y_train)
```

## **AvaliaÃ§Ã£o dos Modelos**

Avalie cada modelo utilizando as seguintes mÃ©tricas:

Algum modelo apresentou overfitting ou underfitting? Se sim, considere as possÃ­veis causas e ajustes:

Overfitting: Quando o modelo aprende demais sobre os dados de treino, perdendo a capacidade de generalizar para novos dados. Considere reduzir a complexidade do modelo ou aumentar os dados de treino.

### ðŸ“Š **AvaliaÃ§Ã£o dos Modelos: RegressÃ£o LogÃ­stica vs Random Forest**

Vamos analisar o desempenho dos dois modelos com base nas mÃ©tricas solicitadas:

**Modelo 1:** RegressÃ£o LogÃ­stica (com normalizaÃ§Ã£o)

**Modelo 2**: Random Forest (sem normalizaÃ§Ã£o, com hiperparÃ¢metros otimizados)

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report

# RegressÃ£o LogÃ­stica
y_pred_lr = pipeline_lr.predict(X_test)

# Random Forest
y_pred_rf = model_rf.predict(X_test)

# FunÃ§Ã£o auxiliar para imprimir mÃ©tricas
def avaliar_modelo(y_true, y_pred, nome):
    print(f"\nðŸ“‹ {nome} â€” AvaliaÃ§Ã£o:")
    print("AcurÃ¡cia      :", round(accuracy_score(y_true, y_pred), 4))
    print("PrecisÃ£o      :", round(precision_score(y_true, y_pred), 4))
    print("Recall        :", round(recall_score(y_true, y_pred), 4))
    print("F1-score      :", round(f1_score(y_true, y_pred), 4))
    print("Matriz ConfusÃ£o:\n", confusion_matrix(y_true, y_pred))
    print("\nRelatÃ³rio:\n", classification_report(y_true, y_pred))

# Avaliar ambos os modelos
avaliar_modelo(y_test, y_pred_lr, "RegressÃ£o LogÃ­stica")
avaliar_modelo(y_test, y_pred_rf, "Random Forest")
```

ðŸŽ¯ Embora a **Random Forest** tenha melhor precisÃ£o, a **RegressÃ£o LogÃ­stica superou no recall e F1-score**, que sÃ£o essenciais em churn (cancelamento), pois queremos identificar o maior nÃºmero possÃ­vel de evasores sem sobrecarregar com falsos positivos.

### **ðŸ” InterpretaÃ§Ã£o CrÃ­tica**

**Vantagens:**

*   Capacidade de explicar o impacto de variÃ¡veis por meio dos coeficientes

**LimitaÃ§Ãµes:**

**Ajustes sugeridos:**

*   Trocar para modelos mais expressivos como SVM ou Redes Neurais

*   Ajustar hiperparÃ¢metros (C, regularizaÃ§Ã£o)


ðŸŒ²**Random Forest:**

**Vantagens:**

*   Robusto contra outliers e dados desbalanceados

**LimitaÃ§Ãµes:**
