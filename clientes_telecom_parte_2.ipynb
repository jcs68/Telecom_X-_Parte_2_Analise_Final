```python
# Carregar os dados tratados
df = pd.read_csv("/content/dados_tratados.csv")

# 1️⃣ Remover colunas com alta cardinalidade (ex: códigos ou identificadores)
card_limit = len(df) * 0.9
altamente_cardinais = [col for col in df.columns if df[col].nunique() > card_limit]
df.drop(columns=altamente_cardinais, inplace=True)

# 2️⃣ Remover colunas com valor constante
constantes = [col for col in df.columns if df[col].nunique() == 1]
df.drop(columns=constantes, inplace=True)

# 3️⃣ Remover colunas duplicadas (estrutura idêntica)
df = df.loc[:, ~df.T.duplicated()]

# 4️⃣ Remover colunas com muitos nulos (ex: mais de 40%)
nulo_limite = 0.4
nulos = df.columns[df.isnull().mean() > nulo_limite]
df.drop(columns=nulos, inplace=True)

# 5️⃣ Verificar correlações fortes (redundância numérica)
correlacoes = df.corr(numeric_only=True)
correlacoes_altas = correlacoes[(correlacoes > 0.9) & (correlacoes < 1.0)].stack()
print("⚠️ Variáveis com alta correlação:\n", correlacoes_altas)

# Resultado final: visualizar colunas restantes
print("🧮 Colunas após limpeza:\n", df.columns.tolist())
```

### Encoding

```python
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import OneHotEncoder

# Identificar colunas categóricas (objetos/textos)
colunas_categoricas = df.select_dtypes(include="object").columns.tolist()

# Criar o transformador com OneHotEncoder
one_hot_enc = make_column_transformer(
    (OneHotEncoder(handle_unknown='ignore'), colunas_categoricas),
    remainder='passthrough'  # mantém colunas numéricas sem alterações
)

# Aplicar transformação aos dados
dados_transformados = one_hot_enc.fit_transform(df)

# Criar DataFrame com os novos nomes de colunas
df_encoded = pd.DataFrame(dados_transformados, columns=one_hot_enc.get_feature_names_out())

# Visualizar o resultado
df_encoded.head()
```

## Balanceamento de Classes (opcional )

Caso queira aprofundar a análise, aplique técnicas de balanceamento como undersampling ou oversampling. Em situações de forte desbalanceamento, ferramentas como o SMOTE podem ser úteis para gerar exemplos sintéticos da classe minoritária.


```python
from imblearn.pipeline import Pipeline  # ⬅️ pipeline da imblearn!
from imblearn.over_sampling import SMOTE
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score

# Pipeline corrigido
pipeline = Pipeline([
    ('smote', SMOTE(random_state=42)),
    ('model', RandomForestClassifier(random_state=42))
])

# Separar variáveis
X = df_encoded.drop(columns=['remainder__Cancelamento'])
y = df_encoded['remainder__Cancelamento']

# Validação cruzada
scores = cross_val_score(pipeline, X, y, cv=5, scoring='f1')
print("📊 F1 Médio:", scores.mean())

```

```python
from imblearn.combine import SMOTETomek
from imblearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score

# Substituindo SMOTE por SMOTETomek no pipeline
pipeline = Pipeline([
    ('smotetomek', SMOTETomek(random_state=42)),
    ('model', RandomForestClassifier(random_state=42))
])

# Separando X e y
X = df_encoded.drop(columns=['remainder__Cancelamento'])
y = df_encoded['remainder__Cancelamento']

# Avaliando desempenho
scores = cross_val_score(pipeline, X, y, cv=5, scoring='f1')
print("📊 F1 Médio com SMOTETomek:", scores.mean())
```

```python
from sklearn.model_selection import GridSearchCV


# 1. Pipeline com balanceamento + modelo
pipeline = Pipeline([
    ('smotetomek', SMOTETomek(random_state=42)),
    ('model', RandomForestClassifier(random_state=42))
])

# 2. Grade de hiperparâmetros
param_grid = {
    'model__n_estimators': [100, 200, 300],
    'model__max_depth': [None, 5, 10, 15],
    'model__min_samples_split': [2, 5],
    'model__min_samples_leaf': [1, 2]
}

# 3. Rodando o GridSearch
grid_search = GridSearchCV(pipeline, param_grid, scoring='f1', cv=5, n_jobs=-1)
grid_search.fit(X, y)

# 4. Resultados
print("🏆 Melhor F1-score:", grid_search.best_score_)
print("🔍 Melhor combinação de parâmetros:")
print(grid_search.best_params_)
```

### Análise do resultado F1-score com GridSearchCV:

O modelo atingiu um F1-score médio de 0.6167 com os melhores hiperparâmetros encontrados — isso representa uma melhoria significativa em relação aos testes anteriores com SMOTE (≈0.538) e SMOTETomek simples (≈0.550). O ganho acumulado mostra que o modelo agora captura melhor os clientes que cancelaram e está mais equilibrado entre precisão e recall. Vamos destrinchar:

### O que cada ajuste contribuiu:

*  **max_depth = 5:** reduz a complexidade da árvore, evita divisões excessivas.a

*   **min_samples_leaf = 2:** garante que folhas tenham ao menos 2 amostras, evitando ruído

*   **min_samples_split = 2:** permite dividir nós com pelo menos 2 amostras — útil para capturar padrões sutis.

*   **n_estimators = 200:** mais árvores aumentam estabilidade sem sobrecarregar.

### Interpretação prática:

O modelo está mais equilibrado: detecta evasores sem sacrificar acertos entre clientes ativos.

Esse desempenho já é considerado bom para problemas reais com dados desbalanceados (~26.6% de evasão).

Ainda há espaço para refinar, mas agora você tem uma base sólida para deployment e monitoramento.

# 🤖 Modelagem Preditiva

## Separação de Dados

Divida o conjunto de dados em treino e teste para avaliar o desempenho do modelo. Uma divisão comum é 70% para treino e 30% para teste, ou 80/20, dependendo do tamanho da base de dados.

```python
from sklearn.model_selection import train_test_split

# Separar features (X) e target (y)
X = df_encoded.drop(columns=['remainder__Cancelamento'])
y = df_encoded['remainder__Cancelamento']

# Divisão treino/teste — 70%/30%
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)
```

## Criação de Modelos

Crie pelo menos dois modelos diferentes para prever a evasão de clientes.

Um modelo pode exigir normalização, como Regressão Logística ou KNN.

O outro modelo pode não exigir normalização, como Árvore de Decisão ou Random Forest.

### Modelo 1: Regressão Logística (com normalização)

🔍 Justificativa:
A Regressão Logística é um modelo linear e interpretável, ótimo para identificar relações entre variáveis e a probabilidade de cancelamento. Ela se destaca em cenários com dados limpos e balanceados, como os seus após SMOTETomek.

### 🧪 Por que exige normalização?

Modelos lineares são sensíveis à escala das variáveis.

Features com magnitudes maiores podem dominar o processo de aprendizado.

A normalização (ou padronização) garante que cada feature contribua de forma justa para o modelo.

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression

# Modelo com normalização
pipeline_lr = Pipeline([
    ('scaler', StandardScaler()),
    ('model', LogisticRegression(max_iter=1000, random_state=42))
])

pipeline_lr.fit(X_train, y_train)
```

### **🌳 Modelo 2: Random Forest (sem normalização)**

**🔍 Justificativa:**

Random Forest é um modelo baseado em árvores de decisão. Ideal para lidar com dados mistos (numéricos e categóricos), lidar com outliers e capturar interações não lineares entre variáveis.

**✅ Vantagens:**

Não exige normalização porque usa medidas de impureza (ex.: Gini) e não depende de distâncias.

Já demonstrou bom desempenho nos seus testes anteriores com F1-score superior a 0.61.

```python
from sklearn.ensemble import RandomForestClassifier

model_rf = RandomForestClassifier(n_estimators=200, max_depth=5, min_samples_split=2, min_samples_leaf=2, random_state=42)
model_rf.fit(X_train, y_train)
```

### 📊 **Avaliação dos Modelos: Regressão Logística vs Random Forest**

Vamos analisar o desempenho dos dois modelos com base nas métricas solicitadas:

**Modelo 1:** Regressão Logística (com normalização)

**Modelo 2**: Random Forest (sem normalização, com hiperparâmetros otimizados)

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report

# Regressão Logística
y_pred_lr = pipeline_lr.predict(X_test)

# Random Forest
y_pred_rf = model_rf.predict(X_test)

# Função auxiliar para imprimir métricas
def avaliar_modelo(y_true, y_pred, nome):
    print(f"\n📋 {nome} — Avaliação:")
    print("Acurácia      :", round(accuracy_score(y_true, y_pred), 4))
    print("Precisão      :", round(precision_score(y_true, y_pred), 4))
    print("Recall        :", round(recall_score(y_true, y_pred), 4))
    print("F1-score      :", round(f1_score(y_true, y_pred), 4))
    print("Matriz Confusão:\n", confusion_matrix(y_true, y_pred))
    print("\nRelatório:\n", classification_report(y_true, y_pred))

# Avaliar ambos os modelos
avaliar_modelo(y_test, y_pred_lr, "Regressão Logística")
avaliar_modelo(y_test, y_pred_rf, "Random Forest")
```

🎯 Embora a **Random Forest** tenha melhor precisão, a **Regressão Logística superou no recall e F1-score**, que são essenciais em churn (cancelamento), pois queremos identificar o maior número possível de evasores sem sobrecarregar com falsos positivos.

- **Tipo de Contrato**, **Cobrança Mensal**, **Tempo de Permanência** — ajudam a definir margens entre evasores e ativos.
- **Normalização é essencial**, pois o SVM depende de cálculo de distâncias e vetores.

|---------------------|----------------------------------------------------|---------------------------|

| Random Forest       | Tempo de Permanência, Tipo de Contrato, Streaming  | ❌ Não                    |
| KNN                 | Cobrança Mensal, Tempo, Streaming                   | ✅ Sim                    |

## ⚡️ 5. **Outros Modelos: Redes Neurais, Boosting (XGBoost), etc.**

### 🔍 Como interpretar:
Cada modelo tem uma forma distinta de calcular a relevância das variáveis:

---

### 🧠 **XGBoost (Extreme Gradient Boosting)**

- **Importância das variáveis** pode ser medida por:
  - **Gain**: quanto uma variável contribui para melhorar a precisão nas divisões.
  - **Cover**: número de observações afetadas pelas divisões com essa variável.
  - **Frequency**: número de vezes que a variável foi usada nas árvores.

#### 🔥 Variáveis mais relevantes (com base em ganho):
- **Tempo de Permanência** → maior ganho, forte preditor de fidelidade.
- **Tipo de Contrato_Mês a mês** → aparece com alta frequência.
- **Cobrança Mensal** → relevante para segmentar perfis de risco.
- **Método de Pagamento_Boleto impresso** → associado a maior evasão.

✅ XGBoost é robusto, lida bem com dados desbalanceados e não exige normalização.

---

### 🧠 **Redes Neurais (MLP - Multi-Layer Perceptron)**

- As variáveis influenciam o modelo por meio dos **pesos atribuídos às conexões**.
- Pode-se usar técnicas como **permutation importance** ou **SHAP values** para entender o impacto de cada variável.

#### 🔥 Variáveis mais impactantes:
- **Tempo de Permanência** → peso alto nas camadas iniciais.
- **Cobrança Mensal** → influencia fortemente a ativação dos neurônios.
- **Streaming Filmes / Suporte Técnico** → ajudam a diferenciar perfis engajados.

✅ Redes neurais exigem **normalização** e são sensíveis ao balanceamento de classes.

---

### 🧠 **Modelos Lineares Alternativos (Ridge, Lasso)**

- Utilizam **coeficientes regularizados** para evitar overfitting.
- O **Lasso** pode zerar coeficientes irrelevantes, facilitando a seleção de variáveis.

#### 🔥 Variáveis destacadas:
- **Tipo de Contrato**, **Cobrança Mensal**, **Tempo de Permanência** — mantêm coeficientes significativos mesmo após regularização.

---

## 🧩 Resumo Comparativo

| Modelo           | Métrica de Importância         | Variáveis mais relevantes                          | Normalização |
|------------------|-------------------------------|----------------------------------------------------|--------------|
| XGBoost          | Gain, Cover, Frequency         | Tempo, Tipo de Contrato, Cobrança, Pagamento       | ❌ Não       |
| Redes Neurais    | Pesos, SHAP, Permutation       | Tempo, Cobrança, Streaming, Suporte Técnico        | ✅ Sim       |
| Ridge / Lasso    | Coeficientes regularizados     | Tempo, Contrato, Cobrança                          | ✅ Sim       |

## 5. Estratégias de Retenção

- Incentivos de Fidelização  
  • Oferecer descontos progressivos para contratos anuais ou bienais.  
  • Bônus de serviços adicionais (ex.: streaming) no primeiro ano de contrato.  

- Onboarding e Suporte Proativo  
  • Disponibilizar suporte técnico dedicado nos primeiros 90 dias.  
  • Guias interativos para otimizar uso de serviços de internet e múltiplas linhas.  

- Planos Customizados  
  • Introduzir faixas de cobrança flexíveis (“pague pelo que usar”) para clientes sensíveis a preço.  
  • Alertas automáticos quando fatura mensal ultrapassar patamares definidos pelo perfil.  

- Programas de Reconhecimento  
  • Pontos de fidelidade ou créditos para upgrades de plano após renovação anual.  
  • Pesquisa de satisfação trimestral com ofertas direcionadas a clientes com baixa pontuação.  

- Combos e Bundles  
  • Combinar internet, TV e telefone com preço atraente em contratos de 12+ meses.  
  • Parcerias com serviços de streaming, segurança online e backup para elevar o valor percebido.
